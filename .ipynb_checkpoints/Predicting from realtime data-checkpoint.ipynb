```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('records.csv')
```

```python
df
```

```python
order=['src_bytes', 'count', 'service', 'srv_count', 'protocol_type',
       'diff_srv_rate', 'same_srv_rate', 'flag', 'dst_bytes',
       'srv_serror_rate', 'logged_in', 'duration', 'lnum_compromised',
       'wrong_fragment', 'is_guest_login', 'num_failed_logins']
```

```python
df=df[(df.columns)&(order)]
```

```python
df
```

```python
df.info()
```

```python
df=df[order]
```

```python
df
```

```python

df['protocol_type'] = df['protocol_type'].astype('category')
df['service'] = df['service'].astype('category')
df['flag'] = df['flag'].astype('category')
cat_columns = df.select_dtypes(['category']).columns
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
```

```python
df.info()
```

```python
import pickle

trained_model = pickle.load(open('Trained_Model.sav', 'rb'))

data=df.to_numpy()
```

```python
result = trained_model.predict(data)
```

```python
result
```

```python
for i in result:
    if i == 0:
        print('Normal')
    if i == 1:
        print('DoS')
    if i == 2:
        print('Probe')
    if i == 3:
        print('R2L')
    if i == 4:
        print('U2R')
```
